### Nhập dữ liệu
Để nhập dữ liệu vào chương trình python, chúng ta sử dụng hàm input()

```python
a = int(input("Nhap chieu dai: ")) # type casting ( int(), float(), bool(), str())
print("Nhap chieu dai la:",a, type(a))
```

#### Thực hành: Viết chương trình nhập vào chiều dài và chiều rộng của hình chữ nhật, sau đó in ra chu vi và diện tích của hình chữ nhật đó

```python
# Nhập các kích thước của hình chữ nhật
a = int(input("Nhap chieu dai: "))
print("Nhap chieu dai la:", a)
b = int(input("Nhap chieu rong: "))
print("Nhap chieu dai la:", b)
```

```python
# Tính toán chu vi, diện tích
p = (a+b) * 2
s = a * b
```

```python
# In kết quả ra màn hình bằng 3 cách
print ("chu vi hinh chu nhat la: ",p)
print ("dien tich hinh chu nhat la: ",s)
```
